```python
Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs to the corresponding view functions. In other words, it is a mechanism for directing incoming HTTP requests to the appropriate handler function in the Flask application.

App routing is essential because it allows developers to define how the application should respond to various URLs or routes. This enables the Flask application to be more flexible, as it can handle different types of requests and provide different responses based on the request type and the specific URL.

For example, an e-commerce website may have different routes for different pages such as the homepage, the product listing page, and the product details page. By using app routing in Flask, the application can determine which view function to call based on the URL requested by the user.

In Flask, app routes are defined using the @app.route() decorator. The decorator takes a URL pattern as its argument and associates it with a view function. The view function is then responsible for generating the HTTP response that will be returned to the user.
```

```python
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

The url_for() function is used in Flask for URL building. It is a helper function that generates a URL to a specific function or endpoint in the Flask application. The url_for() function takes the name of the view function or endpoint as its first argument, and any additional arguments are used to generate any necessary URL parameters.

Here's an example that demonstrates the working of the url_for() function:

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/about')
def about():
    return 'About us'

@app.route('/user/<username>')
def user_profile(username):
    return f'Profile page for {username}'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('index'))
        print(url_for('about'))
        print(url_for('user_profile', username='john'))
```
